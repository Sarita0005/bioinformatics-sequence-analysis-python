```python
from google.colab import files
uploaded = files.upload()

path = '/content/gene.fna'
sequence = ''
header = ''

with open (path, 'r') as f:
     for line in f:
         if line.startswith('>'):
            header =line.strip()
         else:
             sequence += line.strip()
print("Header:")
print(header)

print("\nSequence length:")
print(len(sequence))
```

```python
# gc count percent
gc_count = 0
total_bases = 0

for base in sequence.upper():
    if base in ['A','T','G','C']:
       total_bases += 1
    if base == 'C' or base == 'G':
       gc_count += 1

gc_percent = (gc_count/ total_bases)*100

print('GC Percent:', round(gc_percent,2))
```

```python
a_count =0
c_count =0
t_count =0
g_count =0

for base in sequence. upper():
    if base == 'A':
       a_count += 1
    elif base == 'C':
        c_count += 1
    elif base == 'T':
        t_count += 1
    elif base == 'G':
        g_count += 1
total= a_count + g_count + t_count + c_count

gc_percent = (g_count + c_count)/total *100
at_percent = (a_count + t_count)/total *100
gc_at_ratio = (g_count + c_count)/ (a_count + t_count)

print('A:', a_count)
print('T:', t_count)
print('G:', g_count)
print('C:', c_count)

print('GC Percent:', round(gc_percent,2))

print('AT Percent:', round(at_percent,2))
print('gc_at_ratio:', round(gc_at_ratio,3))
```

```python
import pandas as pd
data = {
    'Metric':[
        'Sequence length',
        'A Count',
        'T Count',
        'C Count',
        'G Count',
        'GC Percent',
        'AT Percent',
        'GC/ AT ratio'
    ],

    'Values':[
        total,
        a_count,
        t_count,
        c_count,
        g_count,
        round(gc_percent,2),
        round(at_percent,2),
        round(gc_at_ratio,3),

    ]
}

df = pd.DataFrame(data)
print(df)

# STEP 5: Save results to CSV

df.to_csv("fasta_sequence_summary.csv", index=False)

print("File saved as fasta_sequence_summary.csv")
```

```python
# Codon count
codons =[]
for i in range(0,len(sequence)-2,3):
     codon = sequence[i:i+3]
     codons.append(codon)

print('Total Codon:',len(codons))
print(codons[:10])
```

```python
from collections import Counter

codon_counts = Counter(codons)

print("Unique codons:", len(codon_counts))
print(codon_counts.most_common(10))
```

```python
codon_to_aa = {
    "TTT": "F", "TTC": "F",
    "TTA": "L", "TTG": "L",
    "CTT": "L", "CTC": "L", "CTA": "L", "CTG": "L",
    "ATT": "I", "ATC": "I", "ATA": "I",
    "ATG": "M",
    "GTT": "V", "GTC": "V", "GTA": "V", "GTG": "V",

    "TCT": "S", "TCC": "S", "TCA": "S", "TCG": "S",
    "CCT": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "ACT": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GCT": "A", "GCC": "A", "GCA": "A", "GCG": "A",

    "TAT": "Y", "TAC": "Y",
    "CAT": "H", "CAC": "H",
    "CAA": "Q", "CAG": "Q",
    "AAT": "N", "AAC": "N",
    "AAA": "K", "AAG": "K",
    "GAT": "D", "GAC": "D",
    "GAA": "E", "GAG": "E",

    "TGT": "C", "TGC": "C",
    "TGG": "W",
    "CGT": "R", "CGC": "R", "CGA": "R", "CGG": "R",
    "AGT": "S", "AGC": "S",
    "AGA": "R", "AGG": "R",
    "GGT": "G", "GGC": "G", "GGA": "G", "GGG": "G",

    "TAA": "*", "TAG": "*", "TGA": "*"
}
```

```python
codon_counts_for_aa = codon_counts

aa_count = {}

for codon, count in codon_counts_for_aa.items():
    aa = codon_to_aa.get(codon, "X")  # X = unknown
    if aa in aa_count:
        aa_count[aa] += count
    else:
        aa_count[aa] = count

print("Amino Acid Counts:")
for aa, c in aa_count.items():
    print(aa, ":", c)
```

```python
# First, remove stop codons (*) because they are not real amino acids for visualization.
aa_count_clean = {aa: c for aa, c in aa_count.items() if aa != "*"}
import matplotlib.pyplot as plt

plt.figure()
plt.bar(aa_count_clean.keys(), aa_count_clean.values())
plt.xlabel("Amino Acids")
plt.ylabel("Count")
plt.title("Amino Acid Composition")
plt.show()
```

```python
total_aa = sum(aa_count_clean.values())

aa_freq = {aa: (count / total_aa) * 100
           for aa, count in aa_count_clean.items()}
plt.figure()
plt.bar(aa_freq.keys(), aa_freq.values())
plt.xlabel("Amino Acids")
plt.ylabel("Frequency (%)")
plt.title("Amino Acid Frequency (%)")
plt.show()
import pandas as pd
df_aa = pd.DataFrame({
    "Amino_Acid": aa_freq.keys(),
    "Frequency (%)": aa_freq.values()
})



# sort by frequency
df_aa = df_aa.sort_values(by="Frequency (%)", ascending=False)
print(df_aa)

# save to csv
df_aa.to_csv("amino_acid_frequency.csv", index=False)
print("File saved as amino_acid_frequency.csv")
```
